```python
actDict = {"A":{},"B":{}}
for x in range(0,len(dfEvent)):
  s = dfEvent['ACTIVITY_EN'][x]
  if (s in actDict.get(dfEvent['Automation'][x])):
    actDict.get(dfEvent['Automation'][x]).update({s:actDict.get(dfEvent['Automation'][x]).get(s)+1})
  else:
    actDict.get(dfEvent['Automation'][x]).update({s:1})

for x in actDict:
  for y in actDict.get(x):
    print(x,y,actDict.get(x).get(y))
```

```python
stdDict = {}
for x in range(0,len(dfEvent)):
  cK = str(dfEvent['_CASE_KEY'][x])
  eT = str(dfEvent['EVENTTIME'][x]).split(' ')[1].split(':')
  eTi = int(eT[0])*3600 + int(eT[1])*60 + int(eT[2])
  if (cK in stdDict):
    stdDict.get(cK).append(eTi)
  else:
    stdDict.update({cK:[eTi]})
for x in stdDict:
  print(stdDict.get(x))
```
